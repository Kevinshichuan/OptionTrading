```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# input of variables
d=2
s_0_a=100
s_0_b=100
K=110
r=0.05
sigma_1=0.2
sigma_2=0.3
T = 1
n=100000
rhos = [round(x * 0.1, 1) for x in range(-9, 10)]
y_axis = []
for rho in rhos:
    res = []
    # simulation step
    for i in range(n):
        # Two stocks has the correlation rho
        Z_1 = np.random.normal(0, np.sqrt(T))
        Z_2 = np.random.normal(0, np.sqrt(T))
        s_T_a = s_0_a* np.exp((r-0.5*sigma_1**2)*T+sigma_1*T*Z_1)
        s_T_b = s_0_b* np.exp((r-0.5*sigma_2**2)*T+sigma_2*T*(rho*Z_1+np.sqrt(1-rho**2)*Z_2))
        payoff = max(max(s_T_a,s_T_b)-K,0)

        res.append(payoff*np.exp(0-r*T))

    price = np.mean(res)
    y_axis.append(price)
    
# Plotting the graph
plt.plot(rhos, y_axis)

# Adding labels and title
plt.xlabel('Rho')
plt.ylabel('Option Value')
plt.title('Option Value For rho')

# Displaying the graph
plt.show()
```
